```python
from src.data.vocab import Vocabulary
from src.models.model import ImageCaptionModel
from src.utils.config import VOCAB_FILE, VOCAB_MIN_FREQ, FEATURES_FILE
import torch

# print current directory to verify paths
print("Current directory:", Path().resolve())

# Example usage of the model
VOCAB_FILE = "../data/processed/vocab.json"
vocab = Vocabulary(min_freq=VOCAB_MIN_FREQ).load(VOCAB_FILE)
model = ImageCaptionModel(vocab_size=len(vocab))
print("Model initialized with vocabulary size:", len(vocab))

# EXAMPLE_IMAGE_PATH = "data/raw/images/000000000885.jpg"
FEATURES_FILE = "../data/features/features.pt"
features = torch.load(FEATURES_FILE)
image_features = features[0]  # Assuming the first image's features for testing

model.eval()  # Set model to evaluation mode
caption = model.generate_caption(image_features, vocab)

" ".join(caption)
```
